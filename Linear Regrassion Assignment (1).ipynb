# Data Preperation

There are some categorical variables with two states, they need to be converted into dual state

Categorical variables with more than two states need to be given dummy variables

Deleting car name , car id and company colunm as they don't add to the linear model

```python
file = file.drop('CarName', 1)
```

```python
file = file.drop('car_ID', 1)
file.head()
```

```python
file = file.drop('Company', 1)
file.head()
```

```python
file = file.drop('symboling', 1)
file.head()
```

```python
# List of variables to map

varlist =  ['fueltype']

# Defining the map function

def binary_map1(x):
    return x.map({'diesel': 1, "gas": 0})

# Applying the function to the housing list
file[varlist] = file[varlist].apply(binary_map1)
```

```python
# List of variables to map

varlist =  ['aspiration']

# Defining the map function

def binary_map2(x):
    return x.map({'turbo': 1, "std": 0})

# Applying the function to the housing list
file[varlist] = file[varlist].apply(binary_map2)
```

```python
# List of variables to map

varlist =  ['doornumber']

# Defining the map function

def binary_map3(x):
    return x.map({'four': 1, "two": 0})

# Applying the function to the housing list
file[varlist] = file[varlist].apply(binary_map3)
```

```python
# List of variables to map

varlist =  ['enginelocation']

# Defining the map function

def binary_map4(x):
    return x.map({'rear': 1, "front": 0})

# Applying the function to the housing list
file[varlist] = file[varlist].apply(binary_map4)
```

```python
file.head()
```

# Dummy Variable
Variable having more than two levels need to be given dummy variables

```python
# Get the dummy variables for the feature 'carbody' and store it in a new variable - 'body'
body = pd.get_dummies(file['carbody'])
```

```python
body.head()
```

```python
body = pd.get_dummies(file['carbody'], drop_first = True)
```

```python
body.head()
```

```python
file = pd.concat([file, body], axis = 1)
```

```python
file.head()
```

```python
# Get the dummy variables for the feature 'drivewheel' and store it in a new variable - 'wheel'
wheel = pd.get_dummies(file['drivewheel'], drop_first = True)
```

```python
file = pd.concat([file, wheel], axis = 1)
file.head()
```

```python
# Get the dummy variables for the feature 'enginetype' and store it in a new variable - 'type'
type = pd.get_dummies(file['enginetype'], drop_first = True)
```

```python
file = pd.concat([file, type], axis = 1)
file.head()
```

```python
# Get the dummy variables for the feature 'cylindernumber' and store it in a new variable - 'cyl'
cyl = pd.get_dummies(file['cylindernumber'], drop_first = True)
```

```python
file = pd.concat([file, cyl], axis = 1)
file.head()
```

```python
# Get the dummy variables for the feature 'fuelsystem' and store it in a new variable - 'fuel'
fuel = pd.get_dummies(file['fuelsystem'], drop_first = True)
```

```python
file = pd.concat([file, fuel], axis = 1)
file.head()
```

```python
# Dropping variables for which dummy variables have been added
file.drop(['carbody', 'drivewheel', 'enginetype', 'cylindernumber', 'fuelsystem' ], axis = 1, inplace = True)
```

# Developing Model

Adding all Variables

```python
# Building a linear model
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.params
```

```python
# Print a summary of the linear regression model obtained
print(lr_1.summary())
```

R2 value is pretty high and even adjusted R2 is near by signifying that the model explains large part of data. 
However, this can be a case of overfitting aswell
p values for most data points are significantly high and even data points change sign from .025 to .975 hinting at multicolinearity.

Let's use RFE to bring down variables to 10 and check results

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

There is still strong colinearity present in data. Calculating VIF to check that

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
